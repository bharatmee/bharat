```python
import cv2

# Read the image
img = cv2.imread(r"D:\bharat\Images\sunny1.jpg")

# Convert the image to grayscale
gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the Haar Cascade classifiers for smile and face detection
sm = cv2.CascadeClassifier(r"D:\bharat\Images\haarcascade_smile.xml")
fc = cv2.CascadeClassifier(r"D:\bharat\Images\haarcascade_frontalface_default.xml")

# Detect faces in the image
faces = fc.detectMultiScale(gry, scaleFactor=1.3, minNeighbors=5)

for (x, y, w, h) in faces:
    # Define regions of interest in grayscale and color images
    roi_gry = gry[y:y+h, x:x+w]
    roi_img = img[y:y+h, x:x+w]

    # Detect smiles within the region of interest
    smiles = sm.detectMultiScale(roi_gry, scaleFactor=1.8, minNeighbors=20)

    # Draw rectangles around detected smiles
    for (xs, ys, ws, hs) in smiles:
        cv2.rectangle(roi_img, (xs, ys), (xs+ws, ys+hs), (0, 255, 0), 2)

# Display the image with detected faces and smiles
cv2.imshow("window", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2

# Read the image
img = cv2.imread(r"D:\bharat\Images\sunny1.jpg")

# Convert the image to grayscale
gry = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load the Haar Cascade classifiers for face and eye detection
fc = cv2.CascadeClassifier(r"D:\bharat\Images\haarcascade_frontalface_default.xml")
ey = cv2.CascadeClassifier(r"D:\bharat\Images\haarcascade_eye.xml")

# Detect faces in the image
faces = fc.detectMultiScale(gry, scaleFactor=1.2, minNeighbors=5)

for (x, y, w, h) in faces:
    # Define regions of interest in grayscale and color images
    roi_gry = gry[y:y+h, x:x+w]
    roi_img = img[y:y+h, x:x+w]
    
    # Draw rectangle around the face
    cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
    
    # Detect eyes within the region of interest
    eyes = ey.detectMultiScale(roi_gry, scaleFactor=1.2, minNeighbors=20)
    
    # Draw rectangles around detected eyes
    for (xs, ys, ws, hs) in eyes:
        cv2.rectangle(roi_img, (xs, ys), (xs+ws, ys+hs), (0, 255, 0), 2)

# Display the image with detected faces and eyes
cv2.imshow("window", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
